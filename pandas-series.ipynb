Pandas Series

Importing Pandas

```python
import numpy as np
import pandas as pd
```

Series from lists

```python
#string
country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
#integers
# integers
runs = [13,24,56,78,100]

runs_ser = pd.Series(runs)
```

```python
#custom index
marks =[67,57,89,100]
subject = ['maths','english','science','hindi']
pd.Series(marks,index=subject)
```

```python
#setting a name
marks = pd.Series(marks,index=subject,name='Pradip ke marks')
marks
```

Series from dict

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
marks_series = pd.Series(marks,name='Pradip ke marks')
marks_series
```
